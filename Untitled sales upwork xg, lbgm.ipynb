```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor, StackingRegressor
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score, mean_absolute_percentage_error
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Sales_Data.csv")

# Remove top 1% outliers from ACTUAL
q = df['ACTUAL'].quantile(0.99)
df = df[df['ACTUAL'] <= q]

# Feature Engineering
try:
    df['SALES_WEEK'] = pd.to_datetime(df['SALES_WEEK'], dayfirst=True)
    df['SALE_MONTH'] = df['SALES_WEEK'].dt.month
    df['SALE_DAY'] = df['SALES_WEEK'].dt.day
    df['SALE_DAYOFWEEK'] = df['SALES_WEEK'].dt.dayofweek
    df['SALE_WEEK'] = df['SALES_WEEK'].dt.isocalendar().week
except Exception:
    print("Warning: SALES_WEEK could not be parsed as date. Skipping date-based features.")
    df['SALE_MONTH'] = 0
    df['SALE_DAY'] = 0
    df['SALE_DAYOFWEEK'] = 0
    df['SALE_WEEK'] = 0

# Encode cyclical features
df['WEEK_SIN'] = np.sin(2 * np.pi * df['SALE_WEEK']/52)
df['WEEK_COS'] = np.cos(2 * np.pi * df['SALE_WEEK']/52)

# Ratio Feature: Price per sale
df['PRICE_PER_UNIT'] = df['AVG_UNIT_PRICE'] / (df['ACTUAL'] + 1)

# Binning feature for price
df['PRICE_BUCKET'] = pd.cut(df['AVG_UNIT_PRICE'], bins=[0, 10, 50, 200, df['AVG_UNIT_PRICE'].max()], labels=[0, 1, 2, 3])

# Log transform ACTUAL to reduce skewness
df['LOG_ACTUAL'] = np.log1p(df['ACTUAL'])

# Split data
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Target encoding for high-cardinality categorical variables
high_card_cols = ['PRODUCT_CODE', 'CATEGORY_CODE']
target = 'ACTUAL'

for col in high_card_cols:
    target_mean = train_df.groupby(col)[target].mean()
    train_df[f'{col}_TARGET_ENC'] = train_df[col].map(target_mean)
    test_df[f'{col}_TARGET_ENC'] = test_df[col].map(target_mean)
    test_df[f'{col}_TARGET_ENC'] = test_df[f'{col}_TARGET_ENC'].fillna(train_df[target].mean())
    print(f"âœ… Target encoding applied to: {col}")

# Drop original categorical columns
train_df.drop(columns=high_card_cols, inplace=True)
test_df.drop(columns=high_card_cols, inplace=True)

# Log transformation for AVG_UNIT_PRICE
train_df['LOG_UNIT_PRICE'] = np.log1p(train_df['AVG_UNIT_PRICE'])
test_df['LOG_UNIT_PRICE'] = np.log1p(test_df['AVG_UNIT_PRICE'])

# Define feature set
features = ['PRODUCT_CODE_TARGET_ENC', 'CATEGORY_CODE_TARGET_ENC', 'LOG_UNIT_PRICE',
            'SALE_MONTH', 'SALE_DAY', 'SALE_DAYOFWEEK', 'SALE_WEEK', 'WEEK_SIN', 'WEEK_COS',
            'PRICE_PER_UNIT']

# Target variable
y_train_log = np.log1p(train_df['ACTUAL'])
y_test = test_df['ACTUAL']

X_train = train_df[features]
X_test = test_df[features]

# Train base models
xgb = XGBRegressor(n_estimators=500, learning_rate=0.01, max_depth=6, subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)
lgb = LGBMRegressor(n_estimators=500, learning_rate=0.01, max_depth=6, subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)

# Stacked ensemble
stack = StackingRegressor(
    estimators=[('xgb', xgb), ('lgb', lgb)],
    final_estimator=LinearRegression()
)

stack.fit(X_train, y_train_log)
y_pred_log = stack.predict(X_test)
y_pred = np.expm1(y_pred_log)

# Evaluate
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print("\nðŸ“Š Stacked Model (XGBoost + LightGBM) after Outlier Removal:")
print(f"âœ… MAE  : {mae:.4f}")
print(f"âœ… RMSE : {rmse:.4f}")
print(f"âœ… RÂ²   : {r2:.4f}")
print(f"âœ… MAPE : {mape:.2%}")
```
